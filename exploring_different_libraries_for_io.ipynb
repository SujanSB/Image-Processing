# Image Processing Libraries: A Comparative Overview

Following are the most common libraries used while dealing with image and videos, majorly in computer vision task.

| Library | Primary Use | Data Format | Key Advantage | Limitation |
|---------|-------------|-------------|---------------|------------|
| OpenCV | Real-time CV tasks | BGR (default) | Speed and comprehensive CV tools | Steeper learning curve |
| scikit-image | Scientific analysis | RGB | Advanced algorithms and SciPy integration | Slower for large datasets |
| Matplotlib | Visualization | RGB | Plotting and annotation tools | No image processing capabilities |
| Pillow | Basic manipulation | RGB | Simplicity and format support | Limited to basic operations |


### OpenCV (Open Source Computer Vision Library)


OpenCV (Open Computer Vision) is a robust open-source computer vision and machine learning software library. To get started:

1. Installation is straightforward via pip:
    ```pip install opencv-python```

2. The library provides comprehensive functions for image/video operations. Basic workflow:
    - Read images using `imread()` 
    - Process using built-in functions
    - Display results with `imshow()`
    - Handle window events with `waitKey()` and `destroyAllWindows()`

3. OpenCV uses BGR color format by default, different from RGB used by most other libraries. Color conversion functions are available for format changes.

4. The library supports both Python and C++ interfaces, making it versatile for different development needs.

5. Core functionalities include image/video I/O, processing, analysis, and feature detection, making it ideal for real-time computer vision applications.


````python
import cv2
img = cv2.imread("images/girl1.jpg")  
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
cv2.imshow("Output", gray)
cv2.waitKey(0)
cv2.destroyAllWindows()

# The image is too BI
# Press "ESC" to close the window
````

## scikit-image

Scikit-image (skimage) is a collection of algorithms for image processing built on NumPy and SciPy. Key aspects include:

1. Installation via pip:
    ```pip install scikit-image```

2. Basic workflow involves:
    - Reading images with `io.imread()`
    - Processing with various functions
    - Displaying using `io.imshow()`
    - Saving with `io.imsave()`

3. Works natively with RGB format, making it compatible with other scientific Python libraries

4. Provides advanced algorithms for:
    - Segmentation
    - Geometric transformations
    - Color space manipulation
    - Feature detection
    - Filters and morphology

5. Strong integration with the scientific Python ecosystem makes it ideal for research and analysis tasks


````python

from skimage.io import imsave,imread,imshow

filename = 'images/girl1.jpg'
girl = imread(filename)
imshow(girl)
imsave("gg.png",girl)
````

# Matplotlib and Pillow

Both Matplotlib and Pillow (PIL) serve distinct purposes in image handling:

### Matplotlib
1. Primary purpose is data visualization and plotting
2. Key features for image handling:
    - Displays images via `plt.imshow()`
    - Rich annotation capabilities
    - Multiple color maps support
    - Integration with NumPy arrays
    - Interactive plotting features

### Pillow (PIL - Python Imaging Library)
1. Focuses on image processing tasks:
    - Format conversion
    - Resizing and rotating
    - Basic filtering
    - Drawing shapes/text
    - Color space conversions
2. Installation:
    ```pip install Pillow```
3. Features:
    - Extensive file format support
    - Memory-efficient operations
    - Simple interface for basic manipulations
    - Built-in image enhancement tools
    - Text rendering capabilities


#### **But Understand** :
### **Matplotlib relies on the Pillow library to load image data.**

````python
from PIL import Image

import matplotlib.pyplot as plt

img_rgb = Image.open('images/girl1.jpg')

# Display with Matplotlib
plt.figure(figsize=(10, 5))
plt.subplot(121)
plt.imshow(img_rgb)
plt.title('Load with PIL and Display with Matplotlib')
plt.axis('off')

plt.show()
````
